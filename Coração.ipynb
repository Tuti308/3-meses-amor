```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.sin((np.pi**3) * x) * np.sqrt((np.e**2 - x**2) / 2) + np.sqrt(np.abs(x))

x = np.linspace(-np.e, np.e, 2000)  # valores entre -e e e

# Aplicar a funo
y = f(x)

# Plotar
plt.figure(figsize=(8, 6))
plt.plot(x, y, color="red", linewidth=2)
plt.fill_between(x, y, -2, color="red", alpha=0.3)  # Preencher para dar forma de corao
plt.title("Grfico em forma de corao ", fontsize=16)
plt.axis("off")  # remover os eixos
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# Funo do corao
def f(x):
    return np.sin((np.pi**3) * x) * np.sqrt((np.e**2 - x**2) / 2) + np.sqrt(np.abs(x))

# Intervalo de x
x = np.linspace(-np.e, np.e, 2000)
y = f(x)

# Criar a figura
fig, ax = plt.subplots(figsize=(6, 6))
ax.set_xlim(-np.e, np.e)
ax.set_ylim(min(y)-1, max(y)+1)
ax.axis("off")

line, = ax.plot([], [], color="red", linewidth=2)

# Funo de atualizao (pulsar)
def update(frame):
    scale = 1 + 0.15 * np.sin(2 * np.pi * frame / 30)  # batida
    line.set_data(x, y * scale)
    return line,

# Criar animao
ani = animation.FuncAnimation(fig, update, frames=60, interval=50, blit=True)

plt.close()  # evita duplicar a figura no notebook
from IPython.display import HTML
HTML(ani.to_jshtml())  # Mostra a animao diretamente
```
